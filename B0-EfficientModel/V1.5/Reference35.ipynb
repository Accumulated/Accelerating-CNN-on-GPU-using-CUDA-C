```python
%%capture cap --no-stderr


%%cu

#include <stdio.h>
#include <stdlib.h>
#include <time.h>
#include <math.h>
#include <cuda.h>
#include <cuda_runtime_api.h>
#include "device_launch_parameters.h"
#include <cusolverDn.h>

#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/Input_For_Stem_Layer.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/Stem/Stem_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/functions.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/Input_Matrix.h"

#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MBConv1_0/MBConv1_0_depthwise_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MBConv1_0/MBConv1_0_project_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MBConv1_0/MBConv1_0_squeeze_excitation_parameters.h"

#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_1/MBConv6_1_expansion_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_1/MBConv6_1_depthwise_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_1/MBConv6_1_squeeze_excitation_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_1/MBConv6_1_project_conv_parameters.h"

#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_2/MBConv6_2_expansion_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_2/MBConv6_2_depthwise_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_2/MBConv6_2_squeeze_excitation_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_2/MBConv6_2_project_conv_parameters.h"

#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_3/MBConv6_3_expansion_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_3/MBConv6_3_depthwise_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_3/MBConv6_3_squeeze_excitation_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_3/MBConv6_3_project_conv_parameters.h"

#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_4/MBConv6_4_expansion_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_4/MBConv6_4_depthwise_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_4/MBConv6_4_squeeze_excitation_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_4/MBConv6_4_project_conv_parameters.h"

#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_5/MBConv6_5_expansion_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_5/MBConv6_5_depthwise_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_5/MBConv6_5_squeeze_excitation_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_5/MBConv6_5_project_conv_parameters.h"

#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_6/MBConv6_6_expansion_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_6/MBConv6_6_depthwise_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_6/MBConv6_6_squeeze_excitation_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_6/MBConv6_6_project_conv_parameters.h"

#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_7/MBConv6_7_expansion_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_7/MBConv6_7_depthwise_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_7/MBConv6_7_squeeze_excitation_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_7/MBConv6_7_project_conv_parameters.h"

#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_8/MBConv6_8_expansion_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_8/MBConv6_8_depthwise_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_8/MBConv6_8_squeeze_excitation_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_8/MBConv6_8_project_conv_parameters.h"

#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_9/MBConv6_9_expansion_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_9/MBConv6_9_depthwise_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_9/MBConv6_9_squeeze_excitation_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_9/MBConv6_9_project_conv_parameters.h"

#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_10/MBConv6_10_expansion_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_10/MBConv6_10_depthwise_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_10/MBConv6_10_squeeze_excitation_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_10/MBConv6_10_project_conv_parameters.h"

#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_11/MBConv6_11_expansion_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_11/MBConv6_11_depthwise_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_11/MBConv6_11_squeeze_excitation_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_11/MBConv6_11_project_conv_parameters.h"

#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_12/MBConv6_12_expansion_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_12/MBConv6_12_depthwise_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_12/MBConv6_12_squeeze_excitation_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_12/MBConv6_12_project_conv_parameters.h"

#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_13/MBConv6_13_expansion_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_13/MBConv6_13_depthwise_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_13/MBConv6_13_squeeze_excitation_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_13/MBConv6_13_project_conv_parameters.h"

#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_14/MBConv6_14_expansion_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_14/MBConv6_14_depthwise_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_14/MBConv6_14_squeeze_excitation_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_14/MBConv6_14_project_conv_parameters.h"

#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_15/MBConv6_15_expansion_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_15/MBConv6_15_depthwise_conv_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_15/MBConv6_15_squeeze_excitation_parameters.h"
#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/MbConv6_15/MBConv6_15_project_conv_parameters.h"

#include "/content/drive/MyDrive/Colab_Notebooks/MBCONVS_float/Head/Head_conv_parameters.h"

int show_out = 0;
int MBCONV1_0_flag = 0;

/* Kernel definitions */
__global__ void INPUT_UNROLLING(int stride, int Filter_Height,
                                Matrix Input, Matrix X_unrolled,
                                int Output_Height, int Output_Width)
{  
    int bx = blockIdx.x, by = blockIdx.y, bz = blockIdx.z;
    int tx = threadIdx.x, ty = threadIdx.y;
 
    // Select row and column values 
    int row =  by * TileDW + ty;
    int col = bx * TileDW + tx;
    int depth = bz;
 
    int col_no_strided = col, row_no_strided = row;
    int depth_offset = depth * X_unrolled.width * Filter_Height * Filter_Height;

    /* 
      Note for bx, by and bz= 0, stride = 2: 
          @ tx = 0, ty = 0 -> First multiply the col * stride, row * stride; = 0, 0
                            you are shifting in x direction using local col
                            you are shifting in y direction using local row;
          @ tx = 1, ty = 0 -> First multiply the col * stride, row * stride; = 2, 0 
                            you are shifting in x direction using local col
                            you are shifting in y direction using local row;   
          @ tx = 0, ty = 1 -> First multiply the col * stride, row * stride; = 0, 2 
                            you are shifting in x direction using local col
                            you are shifting in y direction using local row;                   
    */ 
  
    col *= stride; row *= stride;
 
    // Limit number of threads 
    if (row_no_strided < Output_Height && col_no_strided < Output_Width && depth < Input.depth)
    {   
        // Each thread unrolls k x k elements
        for (int local_row = 0; local_row < Filter_Height; local_row++)
        {
            for (int local_col = 0; local_col < Filter_Height; local_col++)
            {                                  
                // 1. local row and column shifts affect the locations in Unrolled matrix
                // 2. For each col and row non strided values -> you are adding an offset to columns and rows in Unrolled matrix
                // 3. Offset the depth using "depth_offset" variable
                X_unrolled.elements[local_col * X_unrolled.width + local_row * Filter_Height * X_unrolled.width + col_no_strided + row_no_strided * Output_Width + depth_offset] = 
                Input.elements[(row + local_row) * Input.width + (col + local_col) + depth * Input.height * Input.width];
            }
        }
    }

}

__global__ void DWConv2d_kernel(Matrix Input, Matrix Filter, Matrix Output, int stride)
{
    int bx = blockIdx.x;
    int by = blockIdx.y;
    int bz = blockIdx.z;

    int tx = threadIdx.x;
    int ty = threadIdx.y;

    int row = by * TileDW + ty;
    int col = bx * TileDW + tx;
    int dep = bz;

    float Pvalue = 0;

    if (row < Output.height && col < Output.width && dep < Output.depth)
    {
        // 1 thread unrolls kxk section
        for (int j = 0; j < Filter.height; j++)
        {
            for (int i = 0; i < Filter.width; i++)
            {
                Pvalue += Filter.elements[j * Filter.width + i + dep * Filter.height * Filter.width] *
                    Input.elements[(j * Input.width + row * stride * Input.width) + (i + col * stride) + dep * Input.height * Input.width];
            }
        }
        Output.elements[row * Output.width + col + dep * Output.height * Output.width] = Pvalue;
    }

}

__global__ void MatrixMulKernel(Matrix M, Matrix N, Matrix P, int num_blocks, int activation, int IS_BIASED, Matrix bias_mat)
{
  __shared__ float Mds[Tile_GEMM][Tile_GEMM];
  __shared__ float Nds[Tile_GEMM][Tile_GEMM];

  int bx = blockIdx.x;
  int by = blockIdx.y;
  int tx = threadIdx.x;
  int ty = threadIdx.y;

  // Identify the row and column of the d_P element to work on
  int Row = by * Tile_GEMM + ty;
  int Col = bx * Tile_GEMM + tx;
  float Pvalue = 0;

  // Loop over the d_M and d_N tiles required to compute d_P element
  for (int ph = 0; ph < num_blocks; ++ph)
  {
    // Collaborative loading of d_M and d_N tiles into shared memory
    if ((Row < M.height) && (ph * Tile_GEMM + tx) < M.width)
    {
      Mds[ty][tx] = M.elements[Row * M.width + ph * Tile_GEMM + tx];
    }

    if ((ph * Tile_GEMM + ty) < N.height && Col < N.width)
    {
      Nds[ty][tx] = N.elements[(ph * Tile_GEMM + ty) * N.width + Col];
    }

    __syncthreads();

    for (int k = 0; k < Tile_GEMM && (ph * Tile_GEMM) + k < M.width; ++k)
    {
      Pvalue += Mds[ty][k] * Nds[k][tx];
    }

    __syncthreads();

  }

  if ((Row < M.height) && (Col < N.width))
  {
    P.elements[Row * P.width + Col] = Pvalue;
    
    switch (IS_BIASED) 
    {
      case BIASED:
        Pvalue = Pvalue + bias_mat.elements[Row];
        break;
      
      default:
        break;
    } 
            
    switch (activation) 
    {
      case SWISH_ACTIVATION:
        // Swish activation function
        P.elements[Row * P.width + Col] = Pvalue / (1 + exp(-1 * Pvalue));
        break;

      case SIGMOID_ACTIVATION:
        // Sigmoid activation function
        P.elements[Row * P.width + Col] = 1 / (1 + exp(-1 * Pvalue));
        break;

      default:
        break;
    }
  
  }

}


__global__ void unroll_Kernel(int stride, int C, int H, int W, int H_out, int W_out, int K, Matrix X, Matrix X_unroll_Trans)
{
    // Note this code was designed to work with 1D blockDim

    int c = 0, s = 0, h_out_local = 0, w_out_local = 0, h_unroll = 0,
        w_base, p, q, Height_tmp = 0, Width_tmp = 0;
    int t = threadIdx.x;
    int W_unroll = H_out * W_out;

    /*
    The total number of threads will be (C * H_out * W_out) as
    1 thread moves kxk section
    */

    // Boundary check to prevent extra threads from working
    if (t < C * W_unroll)
    {
        c = t / W_unroll;
        s = t % W_unroll;
        h_out_local = s / W_out;
        w_out_local = s % W_out;
        h_unroll = h_out_local * W_out + w_out_local;
        Width_tmp = h_unroll;
        w_base = c * K * K;

        if (w_out_local == 0);
        else
            w_out_local *= stride;
        if (h_out_local == 0);
        else
            h_out_local *= stride;

        // 1 thread unrolls kxk section
        for (p = 0; p < K; p++)
        {
            for (q = 0; q < K; q++)
            {
                W_unroll = w_base + p * K + q;
                Height_tmp = W_unroll;

                X_unroll_Trans.elements[Height_tmp * W_out * H_out + Width_tmp] = X.elements[c * X.height * X.width + (h_out_local + p) * X.width + (w_out_local + q)];

            }
        }
    }
}


// Filter unrolling kernel 
__global__ void Filter_unroll_Kernel(int C, int K, int Out_Depth, Matrix Filter, Matrix Filter_unroll)
{
    // 1 thread takes care of 1 feature map
    // C -> 1 filter depth
    // K -> filter width or height, they both are equal
    // Output_Depth -> is just number of filters that exist

    // Threadidx.x takes care of deciding which feature map to work on
    // i.e. threadidx.x = 0 -> work on the 1st feature map to spread
    // threadidx.x = 1 -> work on the 2nd feature map to spread
    // each thread moves kxk blockDim

    int t = threadIdx.x, p = 0, tmp = 0, s = K * K;

    // total = out depth, that gives total number of filters, * 1 Filter's depth 
    int total_number_of_featureMaps_in_Filters = C;

    // Limit number of threads needed
    if (t < total_number_of_featureMaps_in_Filters)
    {
        tmp = t * s;
        for (p = 0; p < s; p++)
        {
            Filter_unroll.elements[tmp + p] = Filter.elements[tmp + p];
        }
    }
}

__global__ void ConvChannelElementWiseMultiplication(Matrix A, Matrix B)
{
    int row = blockIdx.y * Tile_GEMM + threadIdx.y;
    int col = blockIdx.x * Tile_GEMM + threadIdx.x;
    int depth = blockIdx.z;

    int index = depth * A.width * A.height + row * A.width + col;

    if ((row < A.height) && (col < A.width) && (depth < A.depth))
    {
        A.elements[index] = A.elements[index] * B.elements[depth];
    }
}

// Used with MBConv layers that has skip identity = true
__global__ void Identity_Skip(Matrix A, Matrix B)
{
    int row = blockIdx.y * Tile_GEMM + threadIdx.y;
    int col = blockIdx.x * Tile_GEMM + threadIdx.x;
    int depth = blockIdx.z;

    int index = depth * A.width * A.height + row * A.width + col;

    if ((row < A.height) && (col < A.width) && (depth < A.depth))
    {
        A.elements[index] = A.elements[index] + B.elements[index];
    }
}

// 3-D padding H and V
__global__ void Padding_zeros_Kernel(int padding_value, Matrix Original_Padded)
{
    int row = blockIdx.y * Tile_GEMM + threadIdx.y;
    int col = blockIdx.x * Tile_GEMM + threadIdx.x;
    int depth = blockIdx.z;

    int index = depth * Original_Padded.width * Original_Padded.height + row * Original_Padded.width + col;
    float x = 0;

    if ((row < Original_Padded.height) && (col < Original_Padded.width) && (depth < Original_Padded.depth))
    {
        Original_Padded.elements[index] = x;
    }
}

__global__ void Complete_Padding_Process(Matrix Original_Padded, Matrix Original, int padding_value)
{
    // There must be a constant shift between indeces in 2 matrices
    // The code is based on x axis only
    int row = blockIdx.y * blockDim.y + threadIdx.y;
    int col = blockIdx.x * blockDim.x + threadIdx.x;
    int depth = blockIdx.z * blockDim.z + threadIdx.z;

    int index = depth * Original.width * Original.height + row * Original.width + col;
    int Padding_Index = depth * Original_Padded.width * Original_Padded.height + (row + padding_value) * Original_Padded.width + (col + padding_value);

    if ((row < (Original.height)) && (col < (Original.width)) && (depth < (Original.depth)))
    {
        Original_Padded.elements[Padding_Index] = Original.elements[index];
    }
}

/* Batch Normalization Kernels */
const int BLOCK_SIZE = 16;

__global__ void BN_Kernel_Mean_Reduction(Matrix input, Matrix Mean)
{
    /*
        This code works on 2 * Block_Size elements.
        i.e. for 512 Block_Size -> we are reducing 1024 elements.
        Each thread loads 2 elements, one at tx and the
        other shifted by blockIdx.x.
    */

    __shared__ float partialSum[2 * BLOCK_SIZE];
    float tmp = 0;

    unsigned int tx = threadIdx.x;
    int bx = blockDim.x;

    int by_index = blockIdx.y;
    int bx_index = blockIdx.x;

    // The start variable is to get offset for input matrix in loading
    unsigned int start = blockIdx.x * (2 * blockDim.x);
    int start_yDir = blockIdx.y * input.width;

    if (start + tx < input.width && start_yDir < input.height * input.width)
        // Load 2 elements in the shared memory
        partialSum[tx] = input.elements[start + tx + start_yDir];
    else
        partialSum[tx] = tmp;

    if (tx + bx + start < input.width && start_yDir < input.height * input.width)
        partialSum[bx + tx] = input.elements[start + bx + tx + start_yDir];
    else
        partialSum[bx + tx] = tmp;


    unsigned int stride = 0;

    __syncthreads();

    for (stride = blockDim.x; stride > 0; stride = stride / 2)
    {
        __syncthreads();
        if (tx < stride)
            partialSum[tx] += partialSum[tx + stride];
    }

    __syncthreads();


    if (tx == 0)
        Mean.elements[bx_index + by_index * Mean.width] = partialSum[tx];

}

__global__ void ElementWiseSquaring(Matrix A)
{
    int row = blockIdx.y * Tile_GEMM + threadIdx.y;
    int col = blockIdx.x * Tile_GEMM + threadIdx.x;
    int depth = blockIdx.z;

    int index = depth * A.width * A.height + row * A.width + col;

    if ((row < A.height) && (col < A.width) && (depth < A.depth))
    {
        A.elements[index] = A.elements[index] * A.elements[index];
    }
}

__global__ void ElementWiseSubtraction(Matrix A, Matrix mean)
{
    int row = blockIdx.y * Tile_GEMM + threadIdx.y;
    int col = blockIdx.x * Tile_GEMM + threadIdx.x;
    int depth = blockIdx.z;

    int index = depth * A.width * A.height + row * A.width + col;

    if ((row < A.height) && (col < A.width) && (depth < A.depth))
    {
        A.elements[index] = A.elements[index] - mean.elements[depth];
    }
}


__global__ void BN_Kernel_Final_Layer(Matrix A, Matrix mean, Matrix variance, Matrix weight, Matrix bias, int activate)
{
    // Activate values are assigned as follow
    /*
      0 -> no activation, 1 -> swish, 2 -> sigmoid
    */
    int row = blockIdx.y * Tile_GEMM + threadIdx.y;
    int col = blockIdx.x * Tile_GEMM + threadIdx.x;
    int depth = blockIdx.z;

    int index = depth * A.width * A.height + row * A.width + col;
    int index3 = depth;

    float tmp = 0;
 
    if ((row < A.height) && (col < A.width) && (depth < A.depth))
    {
        A.elements[index] = ((A.elements[index] - mean.elements[index3]) / (sqrt(variance.elements[index3] + 0.001))) * weight.elements[index3] + bias.elements[index3];
        tmp = A.elements[index];

        switch (activate) {
                  case 1:
                      // Swish activation function
                      A.elements[index] = tmp / (1 + exp(-1 * tmp));
                      break;
                  case 2:
                      // Sigmoid activation function
                      A.elements[index] = 1 / (1 + exp(-1 * tmp));
                      break;
                  default:
                      break;
                    }
    }
}



int main()
{

    // Filter Matrix details
    Matrix F1, F2, F3, F4, F5;
 
    // Efficient net model:  


    Matrix ConvOutStem; 
    STEM_LAYER(Input_for_stem_conv, Stem_conv2d_weights,
                224, 254, 3,
                3, 3, 3, 32,
                1, 2,
                &ConvOutStem);
  

    printf("\n\nMBConv 1_0 details comming: \n\n"); 
    MBCONV1_0_flag = 1;
    int FD1 = 32, FD2 = 1, FD3 = 8, FD4 = 32, FD5 = 16;
    int DWFilter_size = 3, stride = 1, padding = 1, skip_connection = 0;
   
    // MBConv1_0 layer implementation
    DEFINE_FILTERS_FOR_MBCONV(&F1, NULL, 0, 0, 0 * FD1,
                              &F2, MBConv1_0_depthwise_conv_conv2d_weights, DWFilter_size, DWFilter_size, 32 * FD2,
                              &F3, MBConv1_0_squeeze_excitation1_conv2d_weights, 1, 1, 32 * FD3,
                              &F4, MBConv1_0_squeeze_excitation2_conv2d_weights, 1, 1, 8 * FD4,
                              &F5, MBConv1_0_project_conv_conv2d_weights, 1, 1, 32 * FD5);


    Matrix ConvOut1_0;
    MBConv_Layer(&ConvOutStem, &ConvOut1_0,
                  &F1, &F2, &F3, &F4, &F5,
                  FD1, FD2, FD3, FD4, FD5,
                  ConvOutStem.depth, FD5, DWFilter_size,
                  stride, padding, skip_connection,
                  MBConv1_0_squeeze_excitation1_conv2d_bias, MBConv1_0_squeeze_excitation2_conv2d_bias,
                  NULL, NULL,
                  NULL, NULL,
                  MBConv1_0_depthwise_conv_BN_mean,     MBConv1_0_depthwise_conv_BN_variance,
                  MBConv1_0_depthwise_conv_BN_weights,  MBConv1_0_depthwise_conv_BN_bias,
                  MBConv1_0_project_conv_BN_mean,       MBConv1_0_project_conv_BN_variance,
                  MBConv1_0_project_conv_BN_weights,    MBConv1_0_project_conv_BN_bias);


    MBCONV1_0_flag = 0;


    printf("\n\nMBConv 6_1 details comming: \n\n");
 
    FD1 = 96; FD2 = 1; FD3 = 4; FD4 = 96; FD5 = 24;
    DWFilter_size = 3; stride = 2; padding = 1; skip_connection = 0;
  
    // MBConv6_1 layer implementation
    DEFINE_FILTERS_FOR_MBCONV(&F1, MBConv6_1_expansion_conv_conv2d_weights, 1, 1, 16 * FD1,
                              &F2, MBConv6_1_depthwise_conv_conv2d_weights, DWFilter_size, DWFilter_size, 96 * FD2,
                              &F3, MBConv6_1_squeeze_excitation1_conv2d_weights, 1, 1, 96 * FD3,
                              &F4, MBConv6_1_squeeze_excitation2_conv2d_weights, 1, 1, 4 * FD4,
                              &F5, MBConv6_1_project_conv_conv2d_weights, 1, 1, 96 * FD5);


    Matrix ConvOut;
    MBConv_Layer(&ConvOut1_0, &ConvOut,
                  &F1, &F2, &F3, &F4, &F5,
                  FD1, FD2, FD3, FD4, FD5,
                  ConvOut1_0.depth, FD5, DWFilter_size,
                  stride, padding, skip_connection,
                  MBConv6_1_squeeze_excitation1_conv2d_bias, MBConv6_1_squeeze_excitation2_conv2d_bias,
                  MBConv6_1_expansion_conv_BN_mean, MBConv6_1_expansion_conv_BN_variance,
                  MBConv6_1_expansion_conv_BN_weights, MBConv6_1_expansion_conv_BN_bias,
                  MBConv6_1_depthwise_conv_BN_mean, MBConv6_1_depthwise_conv_BN_variance,
                  MBConv6_1_depthwise_conv_BN_weights, MBConv6_1_depthwise_conv_BN_bias,
                  MBConv6_1_project_conv_BN_mean, MBConv6_1_project_conv_BN_variance,
                  MBConv6_1_project_conv_BN_weights, MBConv6_1_project_conv_BN_bias);

    // MBConv6_2 layer implementation
    printf("\n\nMBConv 2 details comming: \n\n");

    FD1 = 144; FD2 = 1; FD3 = 6; FD4 = 144; FD5 = 24;
    DWFilter_size = 3; stride = 1 ; padding = 1; skip_connection = 1; 
 
    DEFINE_FILTERS_FOR_MBCONV(&F1, MBConv6_2_expansion_conv_conv2d_weights, 1, 1, 24 * FD1,
                              &F2, MBConv6_2_depthwise_conv_conv2d_weights, DWFilter_size, DWFilter_size, 144 * FD2,
                              &F3, MBConv6_2_squeeze_excitation1_conv2d_weights, 1, 1, 144 * FD3,
                              &F4, MBConv6_2_squeeze_excitation2_conv2d_weights, 1, 1, 6 * FD4,
                              &F5, MBConv6_2_project_conv_conv2d_weights, 1, 1, 144 * FD5);

    Matrix ConvOut2;
    MBConv_Layer(&ConvOut, &ConvOut2,
                  &F1, &F2, &F3, &F4, &F5,
                  FD1, FD2, FD3, FD4, FD5,
                  ConvOut.depth, FD5, DWFilter_size,
                  stride, padding, skip_connection,
                  MBConv6_2_squeeze_excitation1_conv2d_bias, MBConv6_2_squeeze_excitation2_conv2d_bias,
                  MBConv6_2_expansion_conv_BN_mean, MBConv6_2_expansion_conv_BN_variance,
                  MBConv6_2_expansion_conv_BN_weights, MBConv6_2_expansion_conv_BN_bias,
                  MBConv6_2_depthwise_conv_BN_mean, MBConv6_2_depthwise_conv_BN_variance,
                  MBConv6_2_depthwise_conv_BN_weights, MBConv6_2_depthwise_conv_BN_bias,
                  MBConv6_2_project_conv_BN_mean, MBConv6_2_project_conv_BN_variance,
                  MBConv6_2_project_conv_BN_weights, MBConv6_2_project_conv_BN_bias);    

    // MBConv6_3 layer implementation
    printf("\n\nMBConv 3 details comming: \n\n");

    FD1 = 144; FD2 = 1; FD3 = 6; FD4 = 144; FD5 = 40;
    DWFilter_size = 5; stride = 2; padding = 2; skip_connection = 0; 
 
    DEFINE_FILTERS_FOR_MBCONV(&F1, MBConv6_3_expansion_conv_conv2d_weights, 1, 1, 24 * FD1,
                              &F2, MBConv6_3_depthwise_conv_conv2d_weights, DWFilter_size, DWFilter_size, 144 * FD2,
                              &F3, MBConv6_3_squeeze_excitation1_conv2d_weights, 1, 1, 144 * FD3,
                              &F4, MBConv6_3_squeeze_excitation2_conv2d_weights, 1, 1, 6 * FD4,
                              &F5, MBConv6_3_project_conv_conv2d_weights, 1, 1, 144 * FD5);

    Matrix ConvOut3;
    MBConv_Layer(&ConvOut2, &ConvOut3,
                  &F1, &F2, &F3, &F4, &F5,
                  FD1, FD2, FD3, FD4, FD5,
                  ConvOut2.depth, FD5, DWFilter_size,
                  stride, padding, skip_connection,
                  MBConv6_3_squeeze_excitation1_conv2d_bias,  MBConv6_3_squeeze_excitation2_conv2d_bias,
                  MBConv6_3_expansion_conv_BN_mean,           MBConv6_3_expansion_conv_BN_variance,
                  MBConv6_3_expansion_conv_BN_weights,        MBConv6_3_expansion_conv_BN_bias,
                  MBConv6_3_depthwise_conv_BN_mean,           MBConv6_3_depthwise_conv_BN_variance,
                  MBConv6_3_depthwise_conv_BN_weights,        MBConv6_3_depthwise_conv_BN_bias,
                  MBConv6_3_project_conv_BN_mean,             MBConv6_3_project_conv_BN_variance,
                  MBConv6_3_project_conv_BN_weights,          MBConv6_3_project_conv_BN_bias);  
 

    // MBConv6_4 layer implementation
    printf("\n\nMBConv 4 details comming: \n\n");

    FD1 = 240; FD2 = 1; FD3 = 10; FD4 = 240; FD5 = 40;
    DWFilter_size = 5; stride = 1; padding = 2; skip_connection = 1; 
 
    DEFINE_FILTERS_FOR_MBCONV(&F1, MBConv6_4_expansion_conv_conv2d_weights, 1, 1, 40 * FD1,
                              &F2, MBConv6_4_depthwise_conv_conv2d_weights, DWFilter_size, DWFilter_size, 240 * FD2,
                              &F3, MBConv6_4_squeeze_excitation1_conv2d_weights, 1, 1, 240 * FD3,
                              &F4, MBConv6_4_squeeze_excitation2_conv2d_weights, 1, 1, 10 * FD4,
                              &F5, MBConv6_4_project_conv_conv2d_weights, 1, 1, 240 * FD5);

    Matrix ConvOut4;
    MBConv_Layer(&ConvOut3, &ConvOut4,
                  &F1, &F2, &F3, &F4, &F5,
                  FD1, FD2, FD3, FD4, FD5,
                  ConvOut3.depth, FD5, DWFilter_size,
                  stride, padding, skip_connection,
                  MBConv6_4_squeeze_excitation1_conv2d_bias,  MBConv6_4_squeeze_excitation2_conv2d_bias,
                  MBConv6_4_expansion_conv_BN_mean,           MBConv6_4_expansion_conv_BN_variance,
                  MBConv6_4_expansion_conv_BN_weights,        MBConv6_4_expansion_conv_BN_bias,
                  MBConv6_4_depthwise_conv_BN_mean,           MBConv6_4_depthwise_conv_BN_variance,
                  MBConv6_4_depthwise_conv_BN_weights,        MBConv6_4_depthwise_conv_BN_bias,
                  MBConv6_4_project_conv_BN_mean,             MBConv6_4_project_conv_BN_variance,
                  MBConv6_4_project_conv_BN_weights,          MBConv6_4_project_conv_BN_bias);   
 
    // MBConv6_5 layer implementation
    printf("\n\nMBConv 5 details comming: \n\n");

    FD1 = 240; FD2 = 1; FD3 = 10; FD4 = 240; FD5 = 80;
    DWFilter_size = 3; stride = 2; padding = 1; skip_connection = 0; 
 
    DEFINE_FILTERS_FOR_MBCONV(&F1, MBConv6_5_expansion_conv_conv2d_weights, 1, 1, 40 * FD1,
                              &F2, MBConv6_5_depthwise_conv_conv2d_weights, DWFilter_size, DWFilter_size, 240 * FD2,
                              &F3, MBConv6_5_squeeze_excitation1_conv2d_weights, 1, 1, 240 * FD3,
                              &F4, MBConv6_5_squeeze_excitation2_conv2d_weights, 1, 1, 10 * FD4,
                              &F5, MBConv6_5_project_conv_conv2d_weights, 1, 1, 240 * FD5);

    Matrix ConvOut5;
    MBConv_Layer(&ConvOut4, &ConvOut5,
                  &F1, &F2, &F3, &F4, &F5,
                  FD1, FD2, FD3, FD4, FD5,
                  ConvOut4.depth, FD5, DWFilter_size,
                  stride, padding, skip_connection,
                  MBConv6_5_squeeze_excitation1_conv2d_bias,  MBConv6_5_squeeze_excitation2_conv2d_bias,
                  MBConv6_5_expansion_conv_BN_mean,           MBConv6_5_expansion_conv_BN_variance,
                  MBConv6_5_expansion_conv_BN_weights,        MBConv6_5_expansion_conv_BN_bias,
                  MBConv6_5_depthwise_conv_BN_mean,           MBConv6_5_depthwise_conv_BN_variance,
                  MBConv6_5_depthwise_conv_BN_weights,        MBConv6_5_depthwise_conv_BN_bias,
                  MBConv6_5_project_conv_BN_mean,             MBConv6_5_project_conv_BN_variance,
                  MBConv6_5_project_conv_BN_weights,          MBConv6_5_project_conv_BN_bias);    

    // MBConv6_6 layer implementation
    printf("\n\nMBConv 6 details comming: \n\n");

    FD1 = 480; FD2 = 1; FD3 = 20; FD4 = 480; FD5 = 80;
    DWFilter_size = 3; stride = 1; padding = 1; skip_connection = 1; 
 
    DEFINE_FILTERS_FOR_MBCONV(&F1, MBConv6_6_expansion_conv_conv2d_weights, 1, 1, 80 * FD1,
                              &F2, MBConv6_6_depthwise_conv_conv2d_weights, DWFilter_size, DWFilter_size, 480 * FD2,
                              &F3, MBConv6_6_squeeze_excitation1_conv2d_weights, 1, 1, 480 * FD3,
                              &F4, MBConv6_6_squeeze_excitation2_conv2d_weights, 1, 1, 20 * FD4,
                              &F5, MBConv6_6_project_conv_conv2d_weights, 1, 1, 480 * FD5);

    Matrix ConvOut6;
    MBConv_Layer(&ConvOut5, &ConvOut6,
                  &F1, &F2, &F3, &F4, &F5,
                  FD1, FD2, FD3, FD4, FD5,
                  ConvOut5.depth, FD5, DWFilter_size,
                  stride, padding, skip_connection,
                  MBConv6_6_squeeze_excitation1_conv2d_bias,  MBConv6_6_squeeze_excitation2_conv2d_bias,
                  MBConv6_6_expansion_conv_BN_mean,           MBConv6_6_expansion_conv_BN_variance,
                  MBConv6_6_expansion_conv_BN_weights,        MBConv6_6_expansion_conv_BN_bias,
                  MBConv6_6_depthwise_conv_BN_mean,           MBConv6_6_depthwise_conv_BN_variance,
                  MBConv6_6_depthwise_conv_BN_weights,        MBConv6_6_depthwise_conv_BN_bias,
                  MBConv6_6_project_conv_BN_mean,             MBConv6_6_project_conv_BN_variance,
                  MBConv6_6_project_conv_BN_weights,          MBConv6_6_project_conv_BN_bias);    

    // MBConv6_7 layer implementation
    printf("\n\nMBConv 7 details comming: \n\n");

    FD1 = 480; FD2 = 1; FD3 = 20; FD4 = 480; FD5 = 80;
    DWFilter_size = 3; stride = 1; padding = 1; skip_connection = 1; 
 
    DEFINE_FILTERS_FOR_MBCONV(&F1, MBConv6_7_expansion_conv_conv2d_weights, 1, 1, 80 * FD1,
                              &F2, MBConv6_7_depthwise_conv_conv2d_weights, DWFilter_size, DWFilter_size, 480 * FD2,
                              &F3, MBConv6_7_squeeze_excitation1_conv2d_weights, 1, 1, 480 * FD3,
                              &F4, MBConv6_7_squeeze_excitation2_conv2d_weights, 1, 1, 20 * FD4,
                              &F5, MBConv6_7_project_conv_conv2d_weights, 1, 1, 480 * FD5);

    Matrix ConvOut7;
    MBConv_Layer(&ConvOut6, &ConvOut7,
                  &F1, &F2, &F3, &F4, &F5,
                  FD1, FD2, FD3, FD4, FD5,
                  ConvOut6.depth, FD5, DWFilter_size,
                  stride, padding, skip_connection,
                  MBConv6_7_squeeze_excitation1_conv2d_bias,  MBConv6_7_squeeze_excitation2_conv2d_bias,
                  MBConv6_7_expansion_conv_BN_mean,           MBConv6_7_expansion_conv_BN_variance,
                  MBConv6_7_expansion_conv_BN_weights,        MBConv6_7_expansion_conv_BN_bias,
                  MBConv6_7_depthwise_conv_BN_mean,           MBConv6_7_depthwise_conv_BN_variance,
                  MBConv6_7_depthwise_conv_BN_weights,        MBConv6_7_depthwise_conv_BN_bias,
                  MBConv6_7_project_conv_BN_mean,             MBConv6_7_project_conv_BN_variance,
                  MBConv6_7_project_conv_BN_weights,          MBConv6_7_project_conv_BN_bias);   

    // MBConv6_8 layer implementation
    printf("\n\nMBConv 8 details comming: \n\n");

    FD1 = 480; FD2 = 1; FD3 = 20; FD4 = 480; FD5 = 112;
    DWFilter_size = 5; stride = 1; padding = 2; skip_connection = 0; 
 
    DEFINE_FILTERS_FOR_MBCONV(&F1, MBConv6_8_expansion_conv_conv2d_weights, 1, 1, 80 * FD1,
                              &F2, MBConv6_8_depthwise_conv_conv2d_weights, DWFilter_size, DWFilter_size, 480 * FD2,
                              &F3, MBConv6_8_squeeze_excitation1_conv2d_weights, 1, 1, 480 * FD3,
                              &F4, MBConv6_8_squeeze_excitation2_conv2d_weights, 1, 1, 20 * FD4,
                              &F5, MBConv6_8_project_conv_conv2d_weights, 1, 1, 480 * FD5);

    Matrix ConvOut8;
    MBConv_Layer(&ConvOut7, &ConvOut8,
                  &F1, &F2, &F3, &F4, &F5,
                  FD1, FD2, FD3, FD4, FD5,
                  ConvOut7.depth, FD5, DWFilter_size,
                  stride, padding, skip_connection,
                  MBConv6_8_squeeze_excitation1_conv2d_bias,  MBConv6_8_squeeze_excitation2_conv2d_bias,
                  MBConv6_8_expansion_conv_BN_mean,           MBConv6_8_expansion_conv_BN_variance,
                  MBConv6_8_expansion_conv_BN_weights,        MBConv6_8_expansion_conv_BN_bias,
                  MBConv6_8_depthwise_conv_BN_mean,           MBConv6_8_depthwise_conv_BN_variance,
                  MBConv6_8_depthwise_conv_BN_weights,        MBConv6_8_depthwise_conv_BN_bias,
                  MBConv6_8_project_conv_BN_mean,             MBConv6_8_project_conv_BN_variance,
                  MBConv6_8_project_conv_BN_weights,          MBConv6_8_project_conv_BN_bias);   

    // MBConv6_9 layer implementation
    printf("\n\nMBConv 9 details comming: \n\n");

    FD1 = 672; FD2 = 1; FD3 = 28; FD4 = 672; FD5 = 112;
    DWFilter_size = 5; stride = 1; padding = 2; skip_connection = 1; 
 
    DEFINE_FILTERS_FOR_MBCONV(&F1, MBConv6_9_expansion_conv_conv2d_weights, 1, 1, 112 * FD1,
                              &F2, MBConv6_9_depthwise_conv_conv2d_weights, DWFilter_size, DWFilter_size, 672 * FD2,
                              &F3, MBConv6_9_squeeze_excitation1_conv2d_weights, 1, 1, 672 * FD3,
                              &F4, MBConv6_9_squeeze_excitation2_conv2d_weights, 1, 1, 28 * FD4,
                              &F5, MBConv6_9_project_conv_conv2d_weights, 1, 1, 672 * FD5);

    Matrix ConvOut9;
    MBConv_Layer(&ConvOut8, &ConvOut9,
                  &F1, &F2, &F3, &F4, &F5,
                  FD1, FD2, FD3, FD4, FD5,
                  ConvOut8.depth, FD5, DWFilter_size,
                  stride, padding, skip_connection,
                  MBConv6_9_squeeze_excitation1_conv2d_bias,  MBConv6_9_squeeze_excitation2_conv2d_bias,
                  MBConv6_9_expansion_conv_BN_mean,           MBConv6_9_expansion_conv_BN_variance,
                  MBConv6_9_expansion_conv_BN_weights,        MBConv6_9_expansion_conv_BN_bias,
                  MBConv6_9_depthwise_conv_BN_mean,           MBConv6_9_depthwise_conv_BN_variance,
                  MBConv6_9_depthwise_conv_BN_weights,        MBConv6_9_depthwise_conv_BN_bias,
                  MBConv6_9_project_conv_BN_mean,             MBConv6_9_project_conv_BN_variance,
                  MBConv6_9_project_conv_BN_weights,          MBConv6_9_project_conv_BN_bias);    
 
    // MBConv6_10 layer implementation
    printf("\n\nMBConv 10 details comming: \n\n");

    FD1 = 672; FD2 = 1; FD3 = 28; FD4 = 672; FD5 = 112;
    DWFilter_size = 5; stride = 1; padding = 2; skip_connection = 1; 
 
    DEFINE_FILTERS_FOR_MBCONV(&F1, MBConv6_10_expansion_conv_conv2d_weights, 1, 1, 112 * FD1,
                              &F2, MBConv6_10_depthwise_conv_conv2d_weights, DWFilter_size, DWFilter_size, 672 * FD2,
                              &F3, MBConv6_10_squeeze_excitation1_conv2d_weights, 1, 1, 672 * FD3,
                              &F4, MBConv6_10_squeeze_excitation2_conv2d_weights, 1, 1, 28 * FD4,
                              &F5, MBConv6_10_project_conv_conv2d_weights, 1, 1, 672 * FD5);

    Matrix ConvOut10;
    MBConv_Layer(&ConvOut9, &ConvOut10,
                  &F1, &F2, &F3, &F4, &F5,
                  FD1, FD2, FD3, FD4, FD5,
                  ConvOut9.depth, FD5, DWFilter_size,
                  stride, padding, skip_connection,
                  MBConv6_10_squeeze_excitation1_conv2d_bias,  MBConv6_10_squeeze_excitation2_conv2d_bias,
                  MBConv6_10_expansion_conv_BN_mean,           MBConv6_10_expansion_conv_BN_variance,
                  MBConv6_10_expansion_conv_BN_weights,        MBConv6_10_expansion_conv_BN_bias,
                  MBConv6_10_depthwise_conv_BN_mean,           MBConv6_10_depthwise_conv_BN_variance,
                  MBConv6_10_depthwise_conv_BN_weights,        MBConv6_10_depthwise_conv_BN_bias,
                  MBConv6_10_project_conv_BN_mean,             MBConv6_10_project_conv_BN_variance,
                  MBConv6_10_project_conv_BN_weights,          MBConv6_10_project_conv_BN_bias);   

    // MBConv6_11 layer implementation
    printf("\n\nMBConv 11 details comming: \n\n");

    FD1 = 672; FD2 = 1; FD3 = 28; FD4 = 672; FD5 = 192;
    DWFilter_size = 5; stride = 2; padding = 2; skip_connection = 0; 
 
    DEFINE_FILTERS_FOR_MBCONV(&F1, MBConv6_11_expansion_conv_conv2d_weights, 1, 1, 112 * FD1,
                              &F2, MBConv6_11_depthwise_conv_conv2d_weights, DWFilter_size, DWFilter_size, 672 * FD2,
                              &F3, MBConv6_11_squeeze_excitation1_conv2d_weights, 1, 1, 672 * FD3,
                              &F4, MBConv6_11_squeeze_excitation2_conv2d_weights, 1, 1, 28 * FD4,
                              &F5, MBConv6_11_project_conv_conv2d_weights, 1, 1, 672 * FD5);

    Matrix ConvOut11;
    MBConv_Layer(&ConvOut10, &ConvOut11,
                  &F1, &F2, &F3, &F4, &F5,
                  FD1, FD2, FD3, FD4, FD5,
                  ConvOut10.depth, FD5, DWFilter_size,
                  stride, padding, skip_connection,
                  MBConv6_11_squeeze_excitation1_conv2d_bias,  MBConv6_11_squeeze_excitation2_conv2d_bias,
                  MBConv6_11_expansion_conv_BN_mean,           MBConv6_11_expansion_conv_BN_variance,
                  MBConv6_11_expansion_conv_BN_weights,        MBConv6_11_expansion_conv_BN_bias,
                  MBConv6_11_depthwise_conv_BN_mean,           MBConv6_11_depthwise_conv_BN_variance,
                  MBConv6_11_depthwise_conv_BN_weights,        MBConv6_11_depthwise_conv_BN_bias,
                  MBConv6_11_project_conv_BN_mean,             MBConv6_11_project_conv_BN_variance,
                  MBConv6_11_project_conv_BN_weights,          MBConv6_11_project_conv_BN_bias);   

    // MBConv6_12 layer implementation
    printf("\n\nMBConv 12 details comming: \n\n");

    FD1 = 1152; FD2 = 1; FD3 = 48; FD4 = 1152; FD5 = 192;
    DWFilter_size = 5; stride = 1; padding = 2; skip_connection = 1; 
 
    DEFINE_FILTERS_FOR_MBCONV(&F1, MBConv6_12_expansion_conv_conv2d_weights, 1, 1, 192 * FD1,
                              &F2, MBConv6_12_depthwise_conv_conv2d_weights, DWFilter_size, DWFilter_size, 1152 * FD2,
                              &F3, MBConv6_12_squeeze_excitation1_conv2d_weights, 1, 1, 1152 * FD3,
                              &F4, MBConv6_12_squeeze_excitation2_conv2d_weights, 1, 1, 48 * FD4,
                              &F5, MBConv6_12_project_conv_conv2d_weights, 1, 1, 1152 * FD5);

    Matrix ConvOut12;
    MBConv_Layer(&ConvOut11, &ConvOut12,
                  &F1, &F2, &F3, &F4, &F5,
                  FD1, FD2, FD3, FD4, FD5,
                  ConvOut11.depth, FD5, DWFilter_size,
                  stride, padding, skip_connection,
                  MBConv6_12_squeeze_excitation1_conv2d_bias,  MBConv6_12_squeeze_excitation2_conv2d_bias,
                  MBConv6_12_expansion_conv_BN_mean,           MBConv6_12_expansion_conv_BN_variance,
                  MBConv6_12_expansion_conv_BN_weights,        MBConv6_12_expansion_conv_BN_bias,
                  MBConv6_12_depthwise_conv_BN_mean,           MBConv6_12_depthwise_conv_BN_variance,
                  MBConv6_12_depthwise_conv_BN_weights,        MBConv6_12_depthwise_conv_BN_bias,
                  MBConv6_12_project_conv_BN_mean,             MBConv6_12_project_conv_BN_variance,
                  MBConv6_12_project_conv_BN_weights,          MBConv6_12_project_conv_BN_bias);   


    // MBConv6_13 layer implementation
    printf("\n\nMBConv 13 details comming: \n\n");

    FD1 = 1152; FD2 = 1; FD3 = 48; FD4 = 1152; FD5 = 192;
    DWFilter_size = 5; stride = 1; padding = 2; skip_connection = 1; 
 
    DEFINE_FILTERS_FOR_MBCONV(&F1, MBConv6_13_expansion_conv_conv2d_weights, 1, 1, 192 * FD1,
                              &F2, MBConv6_13_depthwise_conv_conv2d_weights, DWFilter_size, DWFilter_size, 1152 * FD2,
                              &F3, MBConv6_13_squeeze_excitation1_conv2d_weights, 1, 1, 1152 * FD3,
                              &F4, MBConv6_13_squeeze_excitation2_conv2d_weights, 1, 1, 48 * FD4,
                              &F5, MBConv6_13_project_conv_conv2d_weights, 1, 1, 1152 * FD5);

    Matrix ConvOut13;
    MBConv_Layer(&ConvOut12, &ConvOut13,
                  &F1, &F2, &F3, &F4, &F5,
                  FD1, FD2, FD3, FD4, FD5,
                  ConvOut12.depth, FD5, DWFilter_size,
                  stride, padding, skip_connection,
                  MBConv6_13_squeeze_excitation1_conv2d_bias,  MBConv6_13_squeeze_excitation2_conv2d_bias,
                  MBConv6_13_expansion_conv_BN_mean,           MBConv6_13_expansion_conv_BN_variance,
                  MBConv6_13_expansion_conv_BN_weights,        MBConv6_13_expansion_conv_BN_bias,
                  MBConv6_13_depthwise_conv_BN_mean,           MBConv6_13_depthwise_conv_BN_variance,
                  MBConv6_13_depthwise_conv_BN_weights,        MBConv6_13_depthwise_conv_BN_bias,
                  MBConv6_13_project_conv_BN_mean,             MBConv6_13_project_conv_BN_variance,
                  MBConv6_13_project_conv_BN_weights,          MBConv6_13_project_conv_BN_bias);  


   // MBConv6_14 layer implementation
    printf("\n\nMBConv 14 details comming: \n\n");

    FD1 = 1152; FD2 = 1; FD3 = 48; FD4 = 1152; FD5 = 192;
    DWFilter_size = 5; stride = 1; padding = 2; skip_connection = 1; 
 
    DEFINE_FILTERS_FOR_MBCONV(&F1, MBConv6_14_expansion_conv_conv2d_weights, 1, 1, 192 * FD1,
                              &F2, MBConv6_14_depthwise_conv_conv2d_weights, DWFilter_size, DWFilter_size, 1152 * FD2,
                              &F3, MBConv6_14_squeeze_excitation1_conv2d_weights, 1, 1, 1152 * FD3,
                              &F4, MBConv6_14_squeeze_excitation2_conv2d_weights, 1, 1, 48 * FD4,
                              &F5, MBConv6_14_project_conv_conv2d_weights, 1, 1, 1152 * FD5);

    Matrix ConvOut14;
    MBConv_Layer(&ConvOut13, &ConvOut14,
                  &F1, &F2, &F3, &F4, &F5,
                  FD1, FD2, FD3, FD4, FD5,
                  ConvOut13.depth, FD5, DWFilter_size,
                  stride, padding, skip_connection,
                  MBConv6_14_squeeze_excitation1_conv2d_bias,  MBConv6_14_squeeze_excitation2_conv2d_bias,
                  MBConv6_14_expansion_conv_BN_mean,           MBConv6_14_expansion_conv_BN_variance,
                  MBConv6_14_expansion_conv_BN_weights,        MBConv6_14_expansion_conv_BN_bias,
                  MBConv6_14_depthwise_conv_BN_mean,           MBConv6_14_depthwise_conv_BN_variance,
                  MBConv6_14_depthwise_conv_BN_weights,        MBConv6_14_depthwise_conv_BN_bias,
                  MBConv6_14_project_conv_BN_mean,             MBConv6_14_project_conv_BN_variance,
                  MBConv6_14_project_conv_BN_weights,          MBConv6_14_project_conv_BN_bias);  

   // MBConv6_15 layer implementation
    printf("\n\nMBConv 15 details comming: \n\n");

    FD1 = 1152; FD2 = 1; FD3 = 48; FD4 = 1152; FD5 = 320;
    DWFilter_size = 3; stride = 1; padding = 1; skip_connection = 0; 
 
    DEFINE_FILTERS_FOR_MBCONV(&F1, MBConv6_15_expansion_conv_conv2d_weights, 1, 1, 192 * FD1,
                              &F2, MBConv6_15_depthwise_conv_conv2d_weights, DWFilter_size, DWFilter_size, 1152 * FD2,
                              &F3, MBConv6_15_squeeze_excitation1_conv2d_weights, 1, 1, 1152 * FD3,
                              &F4, MBConv6_15_squeeze_excitation2_conv2d_weights, 1, 1, 48 * FD4,
                              &F5, MBConv6_15_project_conv_conv2d_weights, 1, 1, 1152 * FD5);

    Matrix ConvOut15;
    MBConv_Layer(&ConvOut14, &ConvOut15,
                  &F1, &F2, &F3, &F4, &F5,
                  FD1, FD2, FD3, FD4, FD5,
                  ConvOut14.depth, FD5, DWFilter_size,
                  stride, padding, skip_connection,
                  MBConv6_15_squeeze_excitation1_conv2d_bias,  MBConv6_15_squeeze_excitation2_conv2d_bias,
                  MBConv6_15_expansion_conv_BN_mean,           MBConv6_15_expansion_conv_BN_variance,
                  MBConv6_15_expansion_conv_BN_weights,        MBConv6_15_expansion_conv_BN_bias,
                  MBConv6_15_depthwise_conv_BN_mean,           MBConv6_15_depthwise_conv_BN_variance,
                  MBConv6_15_depthwise_conv_BN_weights,        MBConv6_15_depthwise_conv_BN_bias,
                  MBConv6_15_project_conv_BN_mean,             MBConv6_15_project_conv_BN_variance,
                  MBConv6_15_project_conv_BN_weights,          MBConv6_15_project_conv_BN_bias);    
 
    // Head layer
    Matrix HEAD_OUT;
    HEAD_LAYER(&ConvOut15, Head_conv2d_weights,
                1, 1, 320, 1280,
                0, 1,
                &HEAD_OUT);


}

void DEFINE_FILTERS_FOR_MBCONV(Matrix *f1, float *filter1, int h1, int w1, int dens1,
                               Matrix *f2, float *filter2, int h2, int w2, int dens2,
                               Matrix *f3, float *filter3, int h3, int w3, int dens3,
                               Matrix *f4, float *filter4, int h4, int w4, int dens4,
                               Matrix *f5, float *filter5, int h5, int w5, int dens5)
{
    // Note: No allocations are done, just pointers point to matrices pre-defined
 
    Set_HostMatrix(h1, w1, dens1, f1);
    Set_HostMatrix(h2, w2, dens2, f2);
    Set_HostMatrix(h3, w3, dens3, f3);
    Set_HostMatrix(h4, w4, dens4, f4);
    Set_HostMatrix(h5, w5, dens5, f5);

    f1 -> elements = filter1;
    f2 -> elements = filter2;
    f3 -> elements = filter3;
    f4 -> elements = filter4;
    f5 -> elements = filter5;
    
}

// The last layer in efficient net
void HEAD_LAYER(Matrix *INPUT_MATRIX, float *HEAD_FILTER,
                int filter_height, int filter_width, int filter_depth, int filter_density,
                int padding, int stride,
                Matrix *HEAD_OUT)
{
     
  Matrix F_HEAD;
  F_HEAD.elements = HEAD_FILTER;
  Set_HostMatrix(filter_height, filter_width, filter_depth * filter_density,
                 &F_HEAD);   
       
  // Calculate output dimensions       
  int out_height = (INPUT_MATRIX -> height + 2 * padding - filter_height) / stride + 1;
  int out_width = (INPUT_MATRIX -> width + 2 * padding - filter_width) / stride + 1;
  int out_depth = filter_density;

  // Allow the output from this layer to go accross all other layers       
  set_allocate_Host(HEAD_OUT, out_height, out_width, out_depth);
  
  // 1st 3 layers: Conv2d 1x1: BN: Swish()
  Conv2d_Layer(INPUT_MATRIX,  &F_HEAD, HEAD_OUT,
              stride, padding,
              INPUT_MATRIX -> depth, out_depth, filter_density,
              Conv2d_1_x_1, NO_ACTIVATION,
              0, NULL);
 
  BN_ALL_PRE_DEFINED(HEAD_OUT, SWISH_ACTIVATION, 
                      Head_BN_mean,    Head_BN_variance,
                      Head_BN_weights, Head_BN_bias);

  // 4th layer: Average pooling layer which is just a reduction sum layer
  // Get mean values for all channels; Dims(1 x 1 x InputDepth)
  Matrix MEAN;
  REDUCTION_SUM(HEAD_OUT, &MEAN);
 
  show_out = 1;
  show_me_enhanced(&MEAN, "Average pool layer output::");
  show_out = 0;
 
  // 5th layer: Fully connected layer::
 
  // 1. Set bias matrix
  Matrix bias; bias.elements = Head_linear_bias;
  Set_HostMatrix(1, 1000, 1, &bias);
 
  // 2. Set weights matrix
  Matrix W; W.elements = Head_linear_weights;
  Set_HostMatrix(1280, 1000, 1, &W);
  
  // 3. Set Output matrix details
  Matrix Out1;
  set_allocate_Host(&Out1, 1, 1000, 1);
     
  Conv_vidMultiplier(&Out1, &W, &MEAN,
                      1, 1000, 1,
                      Conv2d_1_x_1, 1,
                      NO_ACTIVATION, 
                      0, NULL);
 
  for (int i = 0; i < 1280; i++)
  {
    Out1.elements[i] += Head_linear_bias[i];
  }

  show_out = 1;
  show_me_enhanced(&Out1, "Model final output::");
  show_out = 0;
}


// The first layer in efficient net
void STEM_LAYER(float *INPUT_IMGAE, float *STEM_FILTER,
                int image_height, int image_width, int image_depth,
                int filter_height, int filter_width, int filter_depth, int filter_density,
                int padding, int stride,
                Matrix *STEM_OUT)
{
  // Input and filter matrices are defined
  Matrix Hinput_Mat;
  Set_HostMatrix(image_height, image_width, image_depth, &Hinput_Mat);
  Hinput_Mat.elements = INPUT_IMGAE;
     
  Matrix F_STEM;
  F_STEM.elements = STEM_FILTER;
  Set_HostMatrix(filter_height, filter_width, filter_depth * filter_density,
                 &F_STEM);   
       
  // Calculate output dimensions       
  int out_height = (image_height + 2 * padding - filter_height) / stride + 1;
  int out_width = (image_width + 2 * padding - filter_width) / stride + 1;
  int out_depth = filter_density;
 
  // Allow the output from this layer to go accross all other layers       
  set_allocate_Host(STEM_OUT, out_height, out_width, out_depth);
 
  Conv2d_Layer(&Hinput_Mat,  &F_STEM, STEM_OUT,
              stride, padding,
              image_depth, out_depth, filter_density,
              Regular_Conv, NO_ACTIVATION,
              0, NULL);
 
  BN_ALL_PRE_DEFINED(STEM_OUT, SWISH_ACTIVATION, 
                      Stem_BN_mean,    Stem_BN_variance,
                      Stem_BN_weights, Stem_BN_bias);
    
}

void REDUCTION_SUM(Matrix* Output_Modified, Matrix *sum)
{
    /*
      The mean will be a row vector of 1 x C;
      where C is number of original matrix channels
    */

    // Define number of blocks in different directions
    int nbx = 0;
    int nby = 0;
    int nbz = 1;

    // Load input Matrix inot device t calculate mean for it
    Matrix DInputMat;
    set_allocate_copy_Matrix_Device_specific(&DInputMat, Output_Modified, "Input allocated for reduction sum",
                                             Output_Modified -> depth, Output_Modified -> height * Output_Modified -> width, 1);

    /* Starting reduction sum calculations */

    // Note: All blocks are 1D threads. We use Block.x to reduce 1 channel elements'
    // Diffferent block.y to address different number of channels
    nbx = (int)ceil((double)DInputMat.width / (2 * BLOCK_SIZE));
    nby = (int)ceil((double)DInputMat.height);

    if (nbx == 0) nbx = 1;
    if (nby == 0) nby = 1;

    // Allocate Mean matrix in device memory
    Matrix DMean;
    Set_DeviceMatrix(nby, nbx, 1, &DMean, "Mean matrix allocated");

    size_t size;
    cudaError err;
 
    // For loop is held to maintain huge number of summations needed
    for (int i = 0; DInputMat.width != 1; i++)
    {
        dim3 dim_Grid2(nbx, nby, nbz);
        dim3 dim_Block2(BLOCK_SIZE, 1, 1);

        // Make sure to synch between multiple runs
        cudaDeviceSynchronize();
        BN_Kernel_Mean_Reduction <<< dim_Grid2, dim_Block2 >>> (DInputMat, DMean);

        // Save and copy mean values array into the filter array
        size = DMean.height * DMean.width * DMean.depth * sizeof(int);
        err = cudaMemcpy(DInputMat.elements, DMean.elements, size, cudaMemcpyDeviceToDevice);
        CheckCudaError("Copying mean matrix elements from ", err);

        // Modify filter width to fit into the new elements width
        DInputMat.width = nbx;
        DInputMat.height = nby;

        // Recalculate number of blocks in x direction
        nbx = (int)ceil((double)DInputMat.width / (2 * BLOCK_SIZE));
        nby = (int)ceil((double)DInputMat.height);

        if (nbx == 0) nbx = 1;
        if (nby == 0) nby = 1;

        // Set width of mean matrix to the new number of blocks
        DMean.width = nbx;
        DMean.height = nby;
    }

    // Set mean matrix to 1 X C X 1 to ease further calculations
    DMean.height = 1; DMean.width = Output_Modified->depth; DMean.depth = 1;

    // Set the results back into the sum matrix to update the pointer
    set_allocate_Host(sum, DMean.height, DMean.width, DMean.depth);
    
    // Copy elements from device to host
    just_copy_DTH(sum, &DMean, "Mean matrix copied back to host");
    
    /* Mean calculation are done for all channels and the result is 1D row vector */

    // Divide by number of pixels in 1 channel
    int Divide_Mean_BY_ME = Output_Modified->height * Output_Modified->width;

    // Divide all elements by Number of pixels
    for (int i = 0; i < DMean.width; i++)
    {
        sum -> elements[i] = sum -> elements[i] / Divide_Mean_BY_ME;
    }
	
    show_me_enhanced(sum, "Reduction values");

    cudaFree(DMean.elements);
  	cudaFree(DInputMat.elements);
}


/*  Squeeze_and_Excite(&tmp2, &SE_OUT, F3, F4,
                      FD4, FD3, FD4, FD3);
                      */
// Note: input and output channels args represents the 2 Conv layers output channels respectively
void Squeeze_and_Excite(Matrix* InputIMG, Matrix* Result,
    Matrix* Filter1, Matrix* Filter2,
    int FilterDensity2, int FilterDensity1,
    int input_channels, int output_channels,
    float * First_bias, float *Second_bias)
{
    /*
       Steps in squeeze and excite layer:
        1. Get mean value for a tensor
        2. pass the mean to the covolution, swish, convolution, sigmoid
        3. the result will be a 1 x 1 x C, multiply elementwise.
          "each element in a channel is multiplied by the result's corresponding channel element"
        
        Filter Density means #filters used
    */

    show_me_enhanced(InputIMG, "Input for squeeze");
    
    // Get mean values for all channels; Dims(1 x 1 x InputDepth)
    Matrix MEAN;
    REDUCTION_SUM(InputIMG, &MEAN);

    // Tmp1 is used as a transition between 2 convolution layers; Dims(1 x 1 x FilterDensity3)
    Matrix tmp1;
    set_allocate_Host(&tmp1, 1, 1, FilterDensity1);
    
    // tmp2 matrix is the result from sigmoid function: Dims(1 x 1 x FilterDensity4)
    Matrix tmp2;
    set_allocate_Host(&tmp2, 1, 1, FilterDensity2);

    // Sequence: Conv1x1, swish, Conv1x1, sigmoid 
    // Warning: Remember to pre-process MEAN matrix to match 1 x 1 x C as it's the input in this case to Conv2d
    Set_HostMatrix(1, 1, InputIMG -> depth, &MEAN);
 
    Conv2d_Layer(&MEAN, Filter1, &tmp1, 1, 0, input_channels, output_channels, FilterDensity1,
                 Conv2d_1_x_1, SWISH_ACTIVATION,
                 BIASED, First_bias);
    
    Conv2d_Layer(&tmp1, Filter2, &tmp2, 1, 0, output_channels, input_channels, FilterDensity2,
                 Conv2d_1_x_1, SIGMOID_ACTIVATION,
                 BIASED, Second_bias);

    // D matrix takes the result from 2 convolutions
    Matrix D;
    Set_DeviceMatrix(tmp2.height, tmp2.width, tmp2.depth, &D,
        "Matrix D represents the result of 2 Convolutions with their activations");
    
    // Copy elements from host to device
    just_copy_HTD(&D, &tmp2, "Copy result from 2 convolutions output to D matrix");
 
    // C matrix takes input image and multiply by the D matrix 
    // The final result is set into C matrix
    Matrix C;
    Set_DeviceMatrix(InputIMG -> height, InputIMG -> width, InputIMG -> depth, &C,
        "Matrix C represents the input to the squeeze");

    // Copy elements from host to device
    just_copy_HTD(&C, InputIMG, "Copy input to C matrix");

    int nbx = (int)ceil((double)C.width / Tile_GEMM);
    int nby = (int)ceil((double)C.height / Tile_GEMM);
    int nbz = C.depth;

    if (nbx == 0) nbx = 1;

    if (nby == 0) nby = 1;

    // This is the only kernel that runs 3d Grid; 
    // Each block in z dimension controls 1 channel  
    dim3 dim_Grid2(nbx, nby, nbz);
    dim3 dim_Block2(Tile_GEMM, Tile_GEMM, 1);

    // C then D, the final multiplication is in C matrix
    ConvChannelElementWiseMultiplication <<< dim_Grid2, dim_Block2 >>> (C, D);
 
    // Result matrix isn't allocated; set and allocate    
    set_allocate_Host(Result, C.height, C.width, C.depth);
 
    // Copy results back to host memory
    just_copy_DTH(Result, &C, "Copy results from C matrix into result pointer");
    
    show_me_enhanced(Result, "Final matrix after squeeze and excitation");

    cudaFree(D.elements);
    cudaFree(C.elements);
    free(tmp1.elements);

}

void Conv2d_Layer(Matrix* InputIMG, Matrix* FilterK, Matrix* ConvOut,
    int stride, int padding,
    int InputChannels, int OutputChannels, int FilterDensity,
    int Conv_Type, int activation_type,
    int BIASED_CHOISE, float *biasMat)
{
    int OutputHeight = 0, OutputWidth = 0, OutputDepth = 0;

    // 1x1 Conv2d is a special case of Convolution
    if (Conv_Type == Conv2d_1_x_1)
    {
        // Conv2d 1x1 has stride = 1, no padding and K = 1

        Matrix* ptr = InputIMG;

        // Input Dimensions is the same as Output dimensions
        // Only Depth of the output channels differ from input
        OutputHeight = ptr -> height; OutputWidth = ptr -> width; OutputDepth = FilterDensity;

        // Modify Filter Matrix to have dimensions ((K^2 * M) x C x 1); K = 1
        Set_HostMatrix(1 * 1 * FilterDensity, ptr->depth, 1, FilterK);

        // Modify Input matrix to have dimensions (C x (H * W) x 1)
        Set_HostMatrix(ptr->depth, ptr->height * ptr->width, 1, InputIMG);

        // Modify Output Matrix preprocessing to have dimesions ((K^2 * M) x (H * W) x 1); K = 1
        Set_HostMatrix(1 * 1 * FilterDensity, OutputWidth * OutputHeight, 1, ConvOut);

        Conv_vidMultiplier(ConvOut, InputIMG, FilterK,
            OutputHeight, OutputWidth, OutputDepth,
            Conv2d_1_x_1, 1,
            activation_type, 
            BIASED_CHOISE, biasMat);
    }
    else if (Conv_Type == DWConv_k_x_k)
    {
        // Ptr is used to alternate between input image and if padding is needed
        Matrix* ptr = InputIMG;

        // DWConv2d has stride = s, padding = p and kernel = k
        OutputHeight = ConvOut -> height; OutputWidth = ConvOut -> width; OutputDepth = ConvOut -> depth;

        Matrix padded_matr;
        if (padding != 0)
        {
            Padding_Zeros_Function(InputIMG, padding, &padded_matr);
            ptr = &padded_matr;
        }
        
        Conv_vidMultiplier(ConvOut, ptr, FilterK,
                            OutputHeight, OutputWidth, OutputDepth,
                            DWConv_k_x_k, stride,
                            activation_type, 
                            BIASED_CHOISE, biasMat);

        // Padded matrix is no longer needed as Convout has the final result
        if (padding != 0)
        {
            free(padded_matr.elements);
        }
    }
    // Any other kernel size goes here
    else
    {
        printf("Regular convolution is the operation in process \n");
        // Regular convolution: Filter and input unrolling
        Matrix* ptr = InputIMG;
        OutputHeight = (ptr -> height + 2 * padding - FilterK -> height) / stride + 1;
        OutputWidth = (ptr -> width + 2 * padding - FilterK -> width) / stride + 1;
        OutputDepth = FilterDensity;

        Matrix padded_matr;
        if (padding != 0)
        {
            Padding_Zeros_Function(InputIMG, padding, &padded_matr);
            ptr = &padded_matr;          
        }

        // 1st phase: Filter unrolling

        // Unrolled filter has dimesnios (M x (C * k * k) x 1)
        Set_HostMatrix(FilterDensity, FilterK -> depth / FilterDensity * FilterK -> height * FilterK -> width,
                      1, FilterK);
        
        show_me_enhanced(FilterK, "Filter unrolled matrix:::: ");
        

        // 2nd phase: Input unrolling

        // The unrolled Input matrix has dimensions((C * k * k) x (H_out * W_out) x 1)
        Matrix INPUT_MODIFIED;
        Set_HostMatrix(ptr -> depth * 3 * 3,
                       OutputHeight * OutputWidth, 1, &INPUT_MODIFIED);
     
        // Temporarly allocate X_transpose matrix
        int Fsize = INPUT_MODIFIED.depth * INPUT_MODIFIED.height * INPUT_MODIFIED.width * sizeof(float);
        INPUT_MODIFIED.elements = (float*) malloc(Fsize);
        printf("Input unrolling details\n\n alocation the following::"
                "Fsize %d  = ptr -> depth %d * FilterK -> height %d * FilterK -> width %d * OutputHeight %d* OutputWidth %d",
               Fsize, ptr -> depth, 3, 3, OutputHeight, OutputWidth);

        Input_Unroll_gpu(stride, ptr, &INPUT_MODIFIED, OutputHeight, OutputWidth, 3);

        // Output Dimensions calculations
        // Convolution output has dimensions of (M x (H_out * W_out) x 1)
        Set_HostMatrix(FilterDensity, OutputWidth * OutputHeight, 1, ConvOut);


        // Perform Multiplication and re-edit the dimensions of output
        Conv_vidMultiplier(ConvOut, &INPUT_MODIFIED, FilterK,
                            OutputHeight, OutputWidth, OutputDepth,
                            Regular_Conv, stride,
                            activation_type,
                            BIASED_CHOISE, biasMat);

        if (padding != 0)
        {
            free(padded_matr.elements);
        }

        //free(X_Trans.elements);
    }
}

// 5 Filters needed to run the 4 layers sequentially
void MBConv_Layer(Matrix* Input, Matrix* MBConvOut,
    Matrix* F1, Matrix* F2, Matrix* F3, Matrix* F4, Matrix* F5,
    int FD1, int FD2, int FD3, int FD4, int FD5,
    int input_channels, int output_channels, int FilterSizeDW,
    int Stride, int padding, int skip,
    float *bias1, float *bias2,
    float *MBConv_expansion_conv_BN_mean,     float *MBConv_expansion_conv_BN_variance,
    float *MBConv_expansion_conv_BN_weights,  float *MBConv_expansion_conv_BN_bias,
    float *MBConv_depthwise_conv_BN_mean,     float *MBConv_depthwise_conv_BN_variance,
    float *MBConv_depthwise_conv_BN_weights,  float *MBConv_depthwise_conv_BN_bias,
    float *MBConv_project_conv_BN_mean,       float *MBConv_project_conv_BN_variance,
    float *MBConv_project_conv_BN_weights,    float *MBConv_project_conv_BN_bias)
{
    // Note: MBConv1_0 doesn't have the expansion conv function
    Matrix tmp1; Matrix *ptr_mat; 
 
    if (MBCONV1_0_flag == 1)
      ptr_mat = Input;
    else
    {
      set_allocate_Host(&tmp1, Input -> height, Input -> width , FD1);
              
      // 1st layer: 1x1 Conv2d, stride = 1, padding = 0, K = 1
      printf("1st layer operation in MBConv: 1x1 Conv2d layer\n\n");
  
      Conv2d_Layer(Input, F1, &tmp1, 1, 0, input_channels, FD1, FD1, Conv2d_1_x_1,
                  NO_ACTIVATION, 0, NULL);
  
      BN_ALL_PRE_DEFINED(&tmp1, SWISH_ACTIVATION, 
                          MBConv_expansion_conv_BN_mean,    MBConv_expansion_conv_BN_variance,
                          MBConv_expansion_conv_BN_weights, MBConv_expansion_conv_BN_bias);
      ptr_mat = &tmp1;
    }

   // 2nd Layer: KxK DWconv, stride = s, padding = p, K = k
    printf("2nd layer operation in MBConv: DWConv layer\n\n");

    // Height and width changes, Only depth remains still
    int OutputHeight = (ptr_mat -> height + 2 * padding - FilterSizeDW)/Stride + 1;
    int OutputWidth = (ptr_mat -> width + 2 * padding - FilterSizeDW)/Stride + 1;
    int OutputDepth = ptr_mat -> depth;
    show_me_enhanced(ptr_mat, "Input to DWConv");
    show_me_enhanced(F2, "Filter of DWConv");
 
    // Set and allocate tmp2 matrix; it's a transistion between expansion and squeeze
    Matrix tmp2;
    set_allocate_Host(&tmp2, OutputHeight, OutputWidth, OutputDepth);
    
    Conv2d_Layer(ptr_mat, F2, &tmp2, Stride, padding, FD1, FD2, FD2, DWConv_k_x_k,
                 NO_ACTIVATION, 0, NULL);
 
    BN_ALL_PRE_DEFINED(&tmp2, SWISH_ACTIVATION, 
                       MBConv_depthwise_conv_BN_mean,     MBConv_depthwise_conv_BN_variance,
                       MBConv_depthwise_conv_BN_weights,  MBConv_depthwise_conv_BN_bias);
 
    // 3rd Layer: squeeze and excitation
    printf("3rd layer operation in MBConv: squeezeExcitation layer\n\n");

    Matrix SE_OUT;
    Squeeze_and_Excite(&tmp2, &SE_OUT, F3, F4,
                        FD4, FD3, FD4, FD3,
                        bias1, bias2);


    // 4th Layer: 1x1 Conv2d
    printf("4th layer operation in MBConv: Conv2d layer\n\n");

    // MBConv output pointer is set and finally updated after this layer execution
    set_allocate_Host(MBConvOut, SE_OUT.height, SE_OUT.width, FD5);
    
    // 1x1 Conv2d layer
    Conv2d_Layer(&SE_OUT, F5, MBConvOut, 1, 0, FD4, FD5, FD5, Conv2d_1_x_1,
                 NO_ACTIVATION, 0, NULL);

    // BatchNorm layer
    BN_ALL_PRE_DEFINED(MBConvOut, NO_ACTIVATION, 
                       MBConv_project_conv_BN_mean,     MBConv_project_conv_BN_variance,
                       MBConv_project_conv_BN_weights,  MBConv_project_conv_BN_bias);
 
    // Skip identity layer
    if(skip)
    {
      printf("There is a skip identity \n");
      MBConv_SKIP_IDENTITY(MBConvOut, Input);
    }
 
    
    show_me_enhanced(MBConvOut, "MBConv final output");
    
    
    if (MBCONV1_0_flag == 0) 
      free(tmp1.elements);
 
    free(tmp2.elements);
    free(SE_OUT.elements);
}

void MBConv_SKIP_IDENTITY(Matrix *parent, Matrix *child)
{
    // D matrix takes the result from 2 convolutions
    Matrix D;
    Set_DeviceMatrix(parent -> height, parent -> width, parent -> depth, &D,
        "Matrix D represents the identity skip output");
    
    // Copy elements from host to device
    just_copy_HTD(&D, parent, "Copy result from parent matrix to D matrix");
 
    // C matrix takes input image and multiply by the D matrix 
    // The final result is set into C matrix
    Matrix C;
    Set_DeviceMatrix(child -> height, child -> width, child -> depth, &C,
        "Matrix C represents the matrix added identity skip output");

    // Copy elements from host to device
    just_copy_HTD(&C, child, "Copy child to C matrix");

    int nbx = (int)ceil((double)D.width / Tile_GEMM);
    int nby = (int)ceil((double)D.height / Tile_GEMM);
    int nbz = D.depth;

    if (nbx == 0) nbx = 1;

    if (nby == 0) nby = 1;

    // This is the only kernel that runs 3d Grid; 
    // Each block in z dimension controls 1 channel  
    dim3 dim_Grid2(nbx, nby, nbz);
    dim3 dim_Block2(Tile_GEMM, Tile_GEMM, 1);

    Identity_Skip <<<dim_Grid2, dim_Block2 >>> (D, C);

    just_copy_DTH(parent, &D, "matrix in Identity skip copied to device");
    //show_me_enhanced(parent, "Matrix output of Identity skip");
}

void BN_ALL_PRE_DEFINED(Matrix* ptr, int activate, float *mean, float *variance, float *weights, float *bias)
{
    /*
      All weights, bias, running mean and running variance
      are pre-defined. Just call the function and use the
      matrices.
      
      All bias, weights, mean and bariance matrices are 1x1xC

      Output Matrix is modified by the equation
      (y = ((x - Mean) / (sqrt(variance) + epsilon)) * weights + bais)
    */

    // 1. Input Matrix is copied into device memory
    Matrix D_input;
    set_allocate_copy_Matrix_Device(&D_input, ptr, "Device input matrix in BN ");


    // 2. Mean Matrix is copied into device memory
    Matrix DWeights;
    set_allocate_copy_array_Device(&DWeights, weights,
                                    1, 1, ptr -> depth,
                                    "Device weights matrix in BN ");
    
    // 3. Variance Matrix is copied into device memory
    Matrix DRunningVariance;
    set_allocate_copy_array_Device(&DRunningVariance, variance,
                                    1, 1, ptr -> depth,
                                    "Device DRunningVariance matrix in BN ");
    

    // 3. Mean Matrix is copied into device memory
    Matrix DRunningMean;
    set_allocate_copy_array_Device(&DRunningMean, mean,
                                    1, 1, ptr -> depth,
                                    "Device DRunningMean matrix in BN ");
    

    // 4. Bias Matrix is copied into device memory
    Matrix DBias;
    set_allocate_copy_array_Device(&DBias, bias,
                                    1, 1, ptr -> depth,
                                    "Device bias matrix in BN ");
      

    int nbx = (int)ceil((double)ptr->width / Tile_GEMM);
    int nby = (int)ceil((double)ptr->height / Tile_GEMM);
    int nbz = ptr->depth;

    if (nbx == 0) nbx = 1;
    if (nby == 0) nby = 1;

    // This is the only kernel that runs 3d Grid; 
    // Each block in z dimension controls 1 channel  
    dim3 dim_Grid3(nbx, nby, nbz);
    dim3 dim_Block3(Tile_GEMM, Tile_GEMM, 1);

    BN_Kernel_Final_Layer <<< dim_Grid3, dim_Block3 >>> (D_input, DRunningMean, DRunningVariance, DWeights, DBias, activate);

    just_copy_DTH(ptr, &D_input, "matrix in BN copied to device");
    show_out = 1;
    show_me_enhanced(ptr, "Matrix output of BN");
    show_out = 0;
 
    cudaFree(D_input.elements);
    cudaFree(DRunningVariance.elements);
    cudaFree(DRunningMean.elements);
    cudaFree(DBias.elements);
}

void Padding_Zeros_Function(Matrix* Original_Matrix_Before, int padding_Value, Matrix* padded_Matrix)
{
    set_allocate_Host(padded_Matrix, 
                          Original_Matrix_Before->height + 2 * padding_Value,
                          Original_Matrix_Before->width + 2 * padding_Value,
                          Original_Matrix_Before->depth);
    
    Matrix D;
    Set_DeviceMatrix(padded_Matrix->height, padded_Matrix->width, padded_Matrix->depth,
                    &D, "padding with whole zeros only");


    int nbx = (int)ceil((double)padded_Matrix->width / Tile_GEMM);
    int nby = (int)ceil((double)padded_Matrix->height / Tile_GEMM);
    int nbz = padded_Matrix -> depth;

    if (nbx == 0) nbx = 1;

    if (nby == 0) nby = 1;

    // This is the only kernel that runs 3d Grid; 
    // Each block in z dimension controls 1 channel  
    dim3 dim_Grid2(nbx, nby, nbz);
    dim3 dim_Block2(Tile_GEMM, Tile_GEMM, 1);

    // Pass to the padding kernels
    Padding_zeros_Kernel <<< dim_Grid2, dim_Block2 >>> (padding_Value, D);

    size_t size = padded_Matrix->width * padded_Matrix->height * padded_Matrix->depth * sizeof(float);
    cudaError err = cudaMemcpy(padded_Matrix->elements, D.elements, size, cudaMemcpyDeviceToHost);
    CheckCudaError("padding for copying back to the host", err);

    // Phase 2; copy to the specific locations
    Matrix D2;
    set_allocate_copy_Matrix_Device(&D2, Original_Matrix_Before,
                                    "conv padding in allocating Output Matrix of the convolution");
    
    Complete_Padding_Process <<< dim_Grid2, dim_Block2 >>> (D, D2, padding_Value);

    // Copy back to host
    just_copy_DTH(padded_Matrix, &D, "padding for copying back to the host");
    
    show_me_enhanced(padded_Matrix, "Padded zeros Matrix");

    cudaFree(D2.elements);
    cudaFree(D.elements);
}


// Call this function directly for 1x1 conv2d. Don't call for DWConv
void Conv_vidMultiplier(Matrix* out_11, Matrix* XXX_Trans, Matrix* Host_Conv_Filter_Mat_Unrolled,
    int ReconstructOutHieght, int ReconstructOutWidth, int ReconstructOutDepth,
    int ConvType, int stride_DW, int activation_type, int BIASED_CHOISE, float *biasMat)
{
    // Load D_CN_IN device memory
    Matrix D_O;
    Set_DeviceMatrix(out_11->height, out_11->width, out_11->depth,
        &D_O, "conv multiplication in allocating Output Matrix of the convolution");


    // Load D_CN_IN device memory
    Matrix D_1;
    set_allocate_copy_Matrix_Device(&D_1, Host_Conv_Filter_Mat_Unrolled, 
                                    "Conv multiplication in copying 1st input elements from host");

    // Load D_CN_IN device memory
    Matrix D_2;
    set_allocate_copy_Matrix_Device(&D_2, XXX_Trans, 
                                    "Conv multiplication in copying 2nd input elements from host");

    // The multiplication kernel is used for the 1x1 Conv2d and kxk Conv2d
    if (ConvType == Conv2d_1_x_1 || ConvType == Regular_Conv)
    {

      printf("Iam sending as input to multiplication: \n\t D_1 with width = %d, height = %d and depth = %d"
            " \n\tD_2 with width = %d, height = %d and depth = %d"
            " \n\tD_Out with width = %d, height = %d and depth = %d \n \n \t\t\t here are the inputs: \n \n",
            D_1.width, D_1.height, D_1.depth,
            D_2.width, D_2.height, D_2.depth,
            D_O.width, D_O.height, D_O.depth);
        show_me_enhanced(Host_Conv_Filter_Mat_Unrolled, "1st input");
        show_me_enhanced(XXX_Trans, "2nd input");

        // Get number of blocks
        int nbx = (int)ceil((double)D_O.width / Tile_GEMM);
        int nby = (int)ceil((double)D_O.height / Tile_GEMM);
        int num_block_for_phases = (int)ceil((double)D_1.width / Tile_GEMM);

        // Check for zero blocks to make sure code runs correctly
        if (nbx == 0) nbx = 1;
        if (nby == 0) nby = 1;

        dim3 dim_Grid2(nbx, nby, 1);
        dim3 dim_Block2(Tile_GEMM, Tile_GEMM, 1);
        printf("nbx = %d, nby = %d, Tile_GEMM = %d, num_block_phases = %d \n", nbx, nby, Tile_GEMM, num_block_for_phases);
     
        // Get ready if bias needed
        Matrix D_BIAS;
        cudaDeviceSynchronize();

        if(BIASED_CHOISE == BIASED)
        {
          set_allocate_copy_array_Device(&D_BIAS, biasMat,
                                          D_O.height, 1, 1, 
                                          "copying Bias to device memory ");  
                 
          // Call shared memory tiled Multiplication  algorithm
          MatrixMulKernel <<< dim_Grid2, dim_Block2 >>> (D_1, D_2, D_O, num_block_for_phases, activation_type, BIASED_CHOISE, D_BIAS);

          cudaFree(D_BIAS.elements);
        }
        else
        {
          MatrixMulKernel <<< dim_Grid2, dim_Block2 >>> (D_1, D_2, D_O, num_block_for_phases, activation_type, BIASED_CHOISE, D_BIAS);            
        }
        cudaDeviceSynchronize();

        cudaError err = cudaGetLastError();        // Get error code

        if ( err != cudaSuccess )
        {
          printf("CUDA Error: %s\n", cudaGetErrorString(err));
          exit(-1);
        }
        else
          printf("Cuda success after calling matrix mul \n");
    }

    // This case is for DWConv2d
    else
    {
       printf("Iam sending as input to multiplication: \n\t D_1 with width = %d, height = %d and depth = %d"
            " \n\tD_2 with width = %d, height = %d and depth = %d"
            " \n\tD_Out with width = %d, height = %d and depth = %d \n \n \t\t\t here are the inputs: \n \n",
            D_1.width, D_1.height, D_1.depth,
            D_2.width, D_2.height, D_2.depth,
            D_O.width, D_O.height, D_O.depth);
             
        int nbx = (int)ceil((double)D_O.width / TileDW);
        int nby = (int)ceil((double)D_O.height / TileDW);
        int nbz = D_O.depth;
        printf("nbx = %d and nby = %d  and nbz = %d\n", nbx, nby, nbz);
        if (nbx == 0) nbx = 1;
        if (nby == 0) nby = 1;

        // This is the only kernel that runs 3d Grid; 
        // Each block in z dimension controls 1 channel  
        dim3 dim_Grid2(nbx, nby, nbz);
        dim3 dim_Block2(TileDW, TileDW, 1);

        DWConv2d_kernel << < dim_Grid2, dim_Block2 >> > (D_2, D_1, D_O, stride_DW);
    }
    printf("Iam here \n");
    // Read C from device memory
    just_copy_DTH(out_11, &D_O, "Copy oyuput after multiplication back to host");
    printf("Iam here \n");
 
    // Reset the output dimensions to continue in the network
    Set_HostMatrix(ReconstructOutHieght, ReconstructOutWidth, ReconstructOutDepth, out_11);
    printf("Output matrix height = %d, width = %d and depth = %d \n",
           out_11 -> height, out_11 -> width, out_11 -> depth);
 
    // Show the output after convolution process
    
    show_me_enhanced(out_11, "Output after conv multiplication");
    
    // Free device memory
    cudaFree(D_2.elements);
    cudaFree(D_1.elements);
    cudaFree(D_O.elements);
}

void Input_Unroll_gpu(int st_stride, Matrix* input, Matrix* X_Trans, int O_H, int O_W, int Filter_Size)
{
    show_me_enhanced(input, "\n 1. Input matrix for input unrolling");
    
    printf("Stride = %d, X_Trans height = %d, width = %d and depth = %d, OH = %d and OW = %d\n"
            "intpu height=  %d, width = %d and depth = %d\n",
           st_stride, X_Trans->height, X_Trans -> width, X_Trans -> depth, O_H, O_W,
           input -> height, input -> width, input -> depth);
 
    // Load input in device memory
    Matrix Device_Input; 
    set_allocate_copy_Matrix_Device(&Device_Input, input,
                                    "input unrolling allocation and copying into a deevice memory");
    
    // Device Input unrolling matrix allocated and ready to be modified
    Matrix Device_Unrolled;
    Set_DeviceMatrix(X_Trans -> height, X_Trans -> width, X_Trans -> depth,
                    &Device_Unrolled, "input unrolling function for transpose");
 

    int nbx = (int)ceil((double)O_W / TileDW);
    int nby = (int)ceil((double)O_H / TileDW);
    int nbz = Device_Input.depth;

    if (nbx == 0) nbx = 1;

    if (nby == 0) nby = 1;
 
    dim3 dim_Grid2(nbx, nby, nbz);
    dim3 dim_Block2(TileDW, TileDW, 1);

    // You need to synch if the kernel isn't working
    cudaDeviceSynchronize(); 
    INPUT_UNROLLING <<< dim_Grid2, dim_Block2 >>> (st_stride, Filter_Size, Device_Input, Device_Unrolled, O_H, O_W);
    cudaDeviceSynchronize(); 
 
    printf("nbx = %d, nby = %d, nbz = %d, TileDW = %d\n\n", nbx, nby, nbz, TileDW);
    printf("Device input -> height = %d, width = %d and depth = %d, \n"
            "Device unrolled -> height = %d, width = %d and depth = %d \n",
           Device_Input.height, Device_Input.width, Device_Input.depth,
           Device_Unrolled.height, Device_Unrolled.width, Device_Unrolled.depth);

    cudaError err = cudaGetLastError();        // Get error code

    if ( err != cudaSuccess )
    {
      printf("CUDA Error: %s\n", cudaGetErrorString(err));
      exit(-1);
    } 


    // Read C from device memory
    size_t size = X_Trans -> width * X_Trans -> height * X_Trans -> depth * sizeof(float);
    err = cudaMemcpy(X_Trans -> elements, Device_Unrolled.elements, size, cudaMemcpyDeviceToHost);
    CheckCudaError("Copying unrolled input back to host", err);
 
    // Copy results back to host memory
    //just_copy_DTH(X_Trans, &Device_Unrolled, "Copying unrolled input back to host");
    
    show_me_enhanced(X_Trans, "Unrolled input:");
    
 
    // Free device memory
    cudaFree(Device_Unrolled.elements);
    cudaFree(Device_Input.elements);
}


// 3 Sequential Operations: Same as "set_allocate_copy_Matrix_Device",
// However, it uses a pointer to float as a parent.
void set_allocate_copy_array_Device(Matrix *child, float *parent,
									int height, int width, int depth,
									char *notification)
{
	Set_DeviceMatrix(height, width, depth, child, notification);

	size_t size = height * width * depth * sizeof(float);
	
	cudaError err = cudaMemcpy(child -> elements, parent, size,
								cudaMemcpyHostToDevice);
	
	CheckCudaError(notification, err);
}

// 3 Sequential Operations: Set dimensions, allocate device memory and copy.
void set_allocate_copy_Matrix_Device(Matrix *child, Matrix *parent, char *notification)
{
	Set_DeviceMatrix(parent -> height, parent -> width, parent -> depth,
					          child, notification);

	size_t size = parent -> height * parent -> width * parent -> depth * sizeof(float);
	
	cudaError err = cudaMemcpy(child -> elements, parent -> elements,
								              size, cudaMemcpyHostToDevice);
								
	CheckCudaError(notification, err);
}

void set_allocate_copy_Matrix_Device_specific(Matrix *child, Matrix *parent, char *notification, int height, int width, int depth)
{
	Set_DeviceMatrix(height, width, depth, child, notification);

	size_t size = child -> height * child -> width * child -> depth * sizeof(float);
	
	cudaError err = cudaMemcpy(child -> elements, parent -> elements,
								            size, cudaMemcpyHostToDevice);
								
	CheckCudaError(notification, err);
}

void just_copy_HTD(Matrix *child, Matrix *parent, char *notification)
{
    // Read C from device memory
  size_t size = parent -> width * parent -> height * parent -> depth * sizeof(float);
    
	cudaError err = cudaMemcpy(child -> elements, parent -> elements, size, cudaMemcpyHostToDevice);
    
	CheckCudaError(notification, err);
}

void just_copy_DTH(Matrix *child, Matrix *parent, char *notification)
{
  // Read C from device memory
  size_t size = parent -> width * parent -> height * parent -> depth * sizeof(float);
    
	cudaError err = cudaMemcpy(child -> elements, parent -> elements, size, cudaMemcpyDeviceToHost);
    
	CheckCudaError(notification, err);
}

void set_allocate_Host(Matrix *ptr, int height, int width, int depth)
{
	// Note this function allocates memory, remember to free 
	Set_HostMatrix(height, width, depth, ptr);
	
	int Fsize = height * width * depth* sizeof(float);
	printf("Fsize = %d: %d * %d * %d", Fsize, height, width, depth);
 
	ptr -> elements = (float *) malloc(Fsize);
}

void FreeHost_Allocated(Matrix *ptr)
{
	free(ptr -> elements);
}

// Allocations for Device matrices
void Set_DeviceMatrix(int height, int width, int depth, Matrix* ptr, char* NamePtr)
{
    ptr->width = width;
    ptr->height = height;
    ptr->depth = depth;

    size_t size = width * height * depth * sizeof(int);
    cudaError err = cudaMalloc(&(ptr->elements), size);
    CheckCudaError(NamePtr, err);
}

void Set_HostMatrix(int height, int width, int depth, Matrix* ptr)
{
    ptr->width = width;
    ptr->height = height;
    ptr->depth = depth;
}

void CheckCudaError(char* ptr, cudaError err)
{
    if (err == cudaSuccess)
        printf("CUDA success in %s: \n", ptr);
    else
        printf("CUDA error in %s: %s\n", ptr, cudaGetErrorString(err));
}

void show_me_enhanced(Matrix* ptr, char* NamePtr)
{
    if(show_out == 1)
    {
      setvbuf(stdout, NULL, _IOLBF, 0);

          printf("%s,"
              "it has height = %d, "
              "width = %d, "
              "depth = %d \n",
              NamePtr, ptr->height, ptr->width, ptr->depth);

          printf("{\n");
          for (int i = 0; i < ptr -> height * ptr -> width * ptr -> depth; i++)
          {
              if (i % ptr->width == 0 && i >= ptr->width)
                  printf("\n");

              if (i % (ptr->width * ptr->height) == 0 && i >= (ptr->width * ptr->height))
                  printf("\n");

              printf("%.8f", ptr->elements[i]);
              if (i + 1 == ptr->height * ptr->width * ptr->depth);
              else
                  printf(", ");
          }

          printf("} \n");
          printf("\n");

          setvbuf(stdout, NULL, _IOLBF, 0);        
    }
    
}
```
